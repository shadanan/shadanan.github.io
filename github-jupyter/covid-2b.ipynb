```python
import pandas as pd
```

# US Covid Cases

```python
cdf = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports_us/07-24-2020.csv')
cdf.head()
```

```python
cdf = cdf.sort_values("Confirmed", ascending=False)
cdf.plot.bar(x="Province_State", y="Confirmed", figsize=(24, 8))
```

California has the majority of Covid cases.

# Positivity Rate

```python
cdf['Positivity_Rate'] = cdf['Confirmed'] / cdf['People_Tested'] * 100
cdf = cdf.sort_values('Positivity_Rate', ascending=False)
cdf.plot.bar(x="Province_State", y="Positivity_Rate", figsize=(24, 8))
```

While California has a high absolute number of cases, their positivity rate is middle of the pack.
